### Задача 1.2

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

**Зависимость кол-ва спален и стоимости домов**

```python
import seaborn as sns

sns.boxplot(x=df['bedrooms'], y=df['price'] / 1000)

plt.title('Взаимосвязь стоимости дома и кол-ва спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость, тыс. у.е.');
```

Вывод: чем больше спален в доме, тем он дороже (за редким исключением). 

**Зависимость грэйда(оценки домов) от стоимости домов**

```python
import numpy as np

xs = df['grade']
ys = df['price'] / 1000
pd.DataFrame(np.array([xs,ys]).T).plot.scatter(0, 1, s=12, grid=True)
plt.title('Взаимосвязь оценки дома и стоимости')
plt.xlabel('Оценка ')
plt.ylabel('Стоимость, в тыс. у.е.')
plt.show();
```

Вывод: для каждого грэйда есть дома в разной ценовой категории, но в среднем можно сказать, что чем выше грэйд, тем выше стоимость дома.

**Матрица корреляции**

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix < 0.3)] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm');
```

Вывод: из матрицы корреляции видно, что наиболее всего стоимость дома зависит от следующих факторов (в порядке убывания): жилой площади, оценки дома, жилой площади ближайщих 15 домов, кол-ва ванных комнат, вида, кол-ва спален, кол-ва этажей, вида на набережную, площади подвала.

**Зависимость кол-ва этажей и стоимости домов**

```python
sns.boxplot(x=df['floors'], y=df['price'] / 1000)

plt.title('Взаимосвязь кол-ва этажей и стоимости домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Стоимость, в тыс. у.е.');
```

Вывод: в основном самыми дорогими являются дома с количествои этажей: 2.5, самыми дешевыми - в основном одноэтажные.

**Зависимость года постройки и стоимости домов**

```python
xs = df['yr_built']
ys = df['price'] / 1000
pd.DataFrame(np.array([xs,ys]).T).plot.scatter(0, 1, s=12, grid=True)
plt.title('Взаимосвязь года постройки и стоимости домов')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость, в тыс. у.е.')
plt.show();
```

Вывод: нет прямой взаимосвязи между годом постройки и стоимости дома. Существуют, как старые, так и новые дома в разной ценовой категории. Например, три самых дорогих дома: 1910, 1940, 2001 годов постройки.
